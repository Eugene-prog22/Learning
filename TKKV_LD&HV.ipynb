```python
import re

slots = {}
current_idx = 15  # Текущая стартовая позиция блока
pattern = r"== Wafer ID: (\d+)"  # Шаблон для извлечения номера слота

while True:
    # Проверка выхода за пределы данных
    if current_idx + 10 >= len(df_data0):
        break

    # Извлекаем номер слота из строки 10 текущего блока (Unnamed: 3)
    wafer_row = df_data0.iloc[current_idx - 5]  # 10-я строка блока (индексация с 0)
    wafer_id_str = str(wafer_row['Unnamed: 3'])

    # Парсим номер слота с помощью регулярного выражения
    match = re.search(pattern, wafer_id_str)
    if not match:
        print(f"Ошибка: не найден Wafer ID в строке {current_idx + 9}")
        current_idx += 1888  # Пропускаем проблемный блок
        continue

    slot_number = int(match.group(1))

    # Извлекаем данные слота (1840 строк)
    end_idx = current_idx + 1840
    slots[slot_number] = df_data0.iloc[current_idx:end_idx].copy()

    # Переходим к следующему блоку (+8 пропущенных строк)
    current_idx = end_idx + 8
```

массив предсталяется в удобном для анализа виде

```python
import numpy as np
for i in df_dict.keys():
  golova = df_dict[i][['Trans','L','W','Vdr']]
  golova = golova.drop_duplicates().reset_index(drop = True)

  exclude_indices = golova.columns
  selected_indicies = [i for i in df_dict[i].columns if i not in exclude_indices and i not in 'Parametr']
  telo = df_dict[i][selected_indicies]

  df = telo


  def reshape_dataframe(df, new_rows, new_cols):
    # Преобразуем в numpy array и вытягиваем в 1D массив
      flat_data = df.values.ravel()

    # Проверка совместимости размеров
      if len(flat_data) != new_rows * new_cols:
        raise ValueError("Невозможно преобразовать: несовместимые размеры")

    # Создаем новый DataFrame с новой структурой
      return pd.DataFrame(
        data=flat_data.reshape(new_rows, new_cols),
        columns=[f'New_Col_{i+1:03d}' for i in range(new_cols)]
    )

# Применяем преобразование
  new_df = reshape_dataframe(df, new_rows= int (len(telo)/len(name_meas)), new_cols= int (len(telo.columns)*len(name_meas)))
  new_df.columns = new_columns

  df_dict[i] = golova.join(new_df)
  df_dict_short[i] = golova.join(new_df[name_meas])
  df_dict_prot[i] = golova.join(new_df)
  #df_dict[i]


```

```python
def safe_conversion(x):
    try:

        return float(x)
    except (ValueError, TypeError):
        return None  # или np.nan
```

```python
Pre_name = df_prot[['Trans']]

Re_number = [i for i in df_prot.columns if i not in Pre_name]
for i in Re_number :
  df_prot[i] = df_prot[i].apply(safe_conversion)
```

```python
columns_to_fill = [
    'UGST [V]', 'Be [A/V2]', 'UGST#01 [V]', 'IDabs [A]',
    'ID [A/um]', 'UBR [V]', 'IBRabs [A]', 'IDSS [A/um]', 'IGSS [A]'
]

# Заполнение пропусков в основных столбцах и их SD
for col in columns_to_fill:
    # Столбец с валидным средним
    valid_avg_col = f'Valid Average_({col})'
    # Заполняем пропуски в основном столбце
    df_prot[col] = df_prot[col].fillna(df_prot[valid_avg_col])

    # Столбец стандартного отклонения (SD)
    sd_col = f'S.D._({col})'
    # Столбец с валидным SD
    valid_sd_col = f'Valid S.D._({col})'
    # Заполняем пропуски в SD столбце
    df_prot[sd_col] = df_prot[sd_col].fillna(df_prot[valid_sd_col])
```

```python
import os
def save_to_excel(df, file_name, sheet_name):
      folder_path = '/content/drive/MyDrive/TKKV-05_Result_Tr_LD&DLD'

      if not os.path.exists(folder_path):
        os.makedirs(folder_path)

      full_path = os.path.join(folder_path, file_name)


      file_exists = os.path.isfile(full_path) # Проверяем существование файла
      mode = 'a' if file_exists else 'w'
      engine = 'openpyxl' if file_exists else None


      with pd.ExcelWriter(full_path, engine=engine, mode=mode) as writer:
        df.to_excel(writer, sheet_name=sheet_name, index=False)

      print(f'Файл сохранён: {full_path}')
```
